```python
"""
Add the bounding boxes for TERN to the pickle files
"""

import json
import pickle
import os
```

```python
capdir = '<path to coco annotations>/coco/annotations'
train_json =  os.path.join(capdir, 'captions_train2014.json')
val_json =  os.path.join(capdir, 'captions_val2014.json')
tern_data = '<path to tern data>'
```

```python
train = json.load(open(train_json,'r'))
val = json.load(open(val_json,'r'))
```

```python
images = train['images'] + val['images']
```

```python

image_hw = {}
for i, image in enumerate(images):
    cocoid = image['id']
    height = image['height']
    width = image['width']
    image_hw[cocoid] = {
        'height':height,
        'width':width
    }
```

```python
#tern_train_dataset = pickle.load(open(os.path.join(tern_data, 'train_set_ltd_l_tern.pickle'),'rb'))
#tern_val_dataset = pickle.load(open(os.path.join(tern_data, 'validation_set_ltd_l_tern.pickle'),'rb'))
tern_test_dataset = pickle.load(open(os.path.join(tern_data, 'test_set_ltd_l_tern.pickle'),'rb'))
```

```python
def add_size_to_dataset(tern_dataset):
    for i, key in enumerate(tern_dataset['images'].keys()):
        print(image_hw[tern_dataset['images'][key]['cocoid']])

        tern_dataset['images'][key]['size'] = image_hw[tern_dataset['images'][key]['cocoid']]
    return tern_dataset
```
