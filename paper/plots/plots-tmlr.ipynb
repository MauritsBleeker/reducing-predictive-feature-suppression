```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
multiplier_df = pd.read_csv (r'data/tmlr/multiplier.csv')
loss_df = pd.read_csv(r'data/tmlr/reconstruction_loss.csv')
rsum_df = pd.read_csv(r'data/tmlr/rsum.csv')
contrastive_df = pd.read_csv(r'data/tmlr/contrastive.csv')
contrastive_baseline_df = pd.read_csv(r'data/tmlr/contrastive_baseline.csv')
```

```python
plt.figure(figsize=(5,5))

f = plt.figure()
plt.plot(multiplier_df['Step'].to_numpy(), multiplier_df['swa_coco_ltd_constraint_b005_tl - multiplier'].to_numpy(), label = "bound 0.05",  linewidth=4)
plt.plot(multiplier_df['Step'].to_numpy(), multiplier_df['swa_coco_ltd_constraint_b010_tl - multiplier'].to_numpy(), label = "bound 0.1", linewidth=4)
plt.plot(multiplier_df['Step'].to_numpy(), multiplier_df['swa_coco_ltd_constraint_b015_tl - multiplier'].to_numpy(), label = "bound 0.1", linewidth=4)
plt.plot(multiplier_df['Step'].to_numpy(), multiplier_df['swa_coco_ltd_constraint_tl - multiplier'].to_numpy(), label = "bound 0.2", linewidth=4)
plt.plot(multiplier_df['Step'].to_numpy(), multiplier_df['swa_coco_ltd_constraint_b025_tl - multiplier'].to_numpy(), label = "bound 0.3", linewidth=4)

plt.title('Trajectory Lambda',  fontsize=20)
plt.xlabel('Training Step',  fontsize=20)
plt.ylabel('Lambda', fontsize=20)
plt.yscale('log')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.ylim([0.01,1000]) 
plt.xlim([0.0,multiplier_df['Step'].to_numpy().max()])  
plt.xticks(size = 15)
plt.yticks(size = 15)
# plt.legend()
plt.grid(True)

f.savefig("/figures/constraint/tmlr/lambda.pdf", bbox_inches='tight')
```

```python
plt.figure(figsize=(5,5))
f = plt.figure()

plt.plot(loss_df['Step'].to_numpy(), loss_df['swa_coco_ltd_constraint_b005_tl - reconstruction_loss'].to_numpy(), label = "LTD Lag: bound 0.05")
plt.plot(loss_df['Step'].to_numpy(), loss_df['swa_coco_ltd_constraint_b010_tl - reconstruction_loss'].to_numpy(), label = "LTD Lag: bound 0.10")
plt.plot(loss_df['Step'].to_numpy(), loss_df['swa_coco_ltd_constraint_b015_tl - reconstruction_loss'].to_numpy(), label = "LTD Lag: bound 0.15")
plt.plot(loss_df['Step'].to_numpy(), loss_df['swa_coco_ltd_constraint_tl - reconstruction_loss'].to_numpy(), label = "LTD Lag: bound 0.2")
plt.plot(loss_df['Step'].to_numpy(), loss_df['swa_coco_ltd_constraint_b025_tl - reconstruction_loss'].to_numpy(), label = "LTD Lag: bound 0.25")
plt.plot(loss_df['Step'].to_numpy(), loss_df['swa_coco_ltd_tl - reconstruction_loss'].to_numpy(), label = "LTD Dual, beta=1")

plt.title('Reconstruction loss',  fontsize=20)
plt.xlabel('Training Step',  fontsize=20)
plt.ylabel('Cosine distance', fontsize=20)
plt.xlim([0.0,multiplier_df['Step'].to_numpy().max()])  
#plt.legend()
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.ylim([0,0.5])
plt.grid(True)

f.savefig("/figures/constraint/tmlr/reconstruction_loss.pdf", bbox_inches='tight')
```
